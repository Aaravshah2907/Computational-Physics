# Random Number Generator

Import necessary libraries

```python
import matplotlib.pyplot as plt
import numpy as np
```

## LSG

```python
def lsg(seed, N, a=1664525, c=1013904223, m=2**32):
    """Linear Congruential Generator (LCG)"""
    x = seed
    numbers = np.zeros(N)
    for i in range(N):
        x = (a * x + c) % m
        u = x/m
        numbers[i] = u
    return numbers
```

#### Example

```python
rand_nums_1 = lsg(seed=2, N=5000,a=11, c=3, m=2**16)
print(rand_nums_1)
```

## Mean & Variance Test

```python
def mean_variance_calc(random_numbers : np.ndarray):
    """Mean and Variance Test"""
    N = len(random_numbers)
    mean = sum(random_numbers) / N
    sq_sum = sum(x ** 2 for x in random_numbers)/N
    variance = sq_sum - mean**2
    return mean, variance
```

```python
mean_rand1, var_rand1 = mean_variance_calc(rand_nums_1)
print(f"Mean: {mean_rand1}, Variance: {var_rand1}")
print(f"Deviation of Mean from 0.5: {abs(mean_rand1 - 0.5)}")
print(f"Deviation of Variance from 1/12: {abs(var_rand1 - 1/12)}")
```

## Auto-correlation Test

```python
def autocorrelation(random_numbers : np.ndarray, lag: int):
    """Auto-correlation Test"""
    N = len(random_numbers)
    mean, var = mean_variance_calc(random_numbers)
    autocorrelation_list = np.zeros(lag)
    for k in range(lag):
        sum2=0
        for i in range(N-k):
            sum2 += random_numbers[i] * random_numbers[i+k]
        cov = (sum2/N) - mean**2
        ratio = cov / var
        autocorrelation_list[k] = ratio

    plt.plot(autocorrelation_list)
    plt.xlabel("lag (k)")
    plt.ylabel("autocorrelation C(k)")
    plt.show()

    return autocorrelation_list
```

```python
autocorrelation(rand_nums_1, lag=50)
```

## ScatterPlots for Adjacent Pairs

```python
def scatter_plot(random_numbers : np.ndarray):
    """Scatter plot for adjacent pairs"""
    x = random_numbers[:-1]
    y = random_numbers[1:]
    plt.scatter(x, y, s=1)
    plt.xlabel("X(i)")
    plt.ylabel("X(i+1)")
    plt.title("Scatter plot for adjacent pairs")
    plt.show()
```

```python
scatter_plot(rand2)
scatter_plot(rand3)
scatter_plot(rand4)
scatter_plot(rand5)
scatter_plot(randc)
scatter_plot(rand_glibc)
```
